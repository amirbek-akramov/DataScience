```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df=df[df.year>=2010]
df=df[df.transmission!='Other']
df=df[df.fuelType!='Other']
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='price')
plt.show()
```

```python
sns.displot(data=df, x='price', hue='transmission', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price', col='transmission', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price', kind='kde', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price', kde=True, height=5, aspect=2)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.kdeplot(data=df, x='price')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.kdeplot(data=df, y='price')
plt.show()
```

```python
sns.displot(data=df, y='price', height=5, aspect=2)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.kdeplot(data=df, x='price', fill=True) # use 'fill' instead of 'shade'
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.kdeplot(data=df, x='year', y='price', fill=True, hue='fuelType')
plt.show()    # 
```

```python
sns.displot(data=df, x='year', y='price', hue='fuelType', kind="kde", height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x="price", y="year", kind="kde", hue='fuelType', height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x="price", y="year", kind="kde", height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x="price", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both') # setka (major, minor, both) qiymatlarini berib ko'ring
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x="mileage", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both') # setka (major, minor, both) qiymatlarini berib ko'ring
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x="price", hue="fuelType", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.ecdfplot(data=df, x="price", hue="transmission")
plt.grid()
plt.show()
```
