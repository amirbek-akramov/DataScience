```python
 import pandas as pd
 import numpy as np
 import seaborn as sns
 import matplotlib.pyplot as plt
 %matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='year', y='price')
plt.show()
```

```python
df = df[df.year>2000]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='year', y='price')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='mileage', y='price')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='mileage', y='price')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=(df['price'], df['mileage']))
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(data=df, x='mileage', y='price')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='mileage', y='price', hue='transmission')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='mileage', y='price', hue='transmission', style='fuelType', s=100)
plt.show()
```

```python
sns.swarmplot(data=df, x=df['fuelType'], y=df['price'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(data=df, x=df['fuelType'], y=df['price'])
plt.show()
```
