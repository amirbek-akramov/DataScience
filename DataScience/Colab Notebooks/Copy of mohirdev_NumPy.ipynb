#### 19. Tasodifiy sonlardan iborat bo'lgan massiv yarating, so'ng ularni quyida berilgan 4 xil usul yordamida musbat sonlarning butun qismini ajratib oling.

`ðŸ’¡Ishora` %, np.floor, astype, np.trunc

````python
random_numbers = np.random.randn(10)
meth1 = random_numbers[random_numbers > 0]
meth2 = random_numbers[np.floor(random_numbers) > 0]
meth3 = random_numbers[random_numbers.astype(int) > 0]
meth4 = random_numbers[np.trunc(random_numbers) > 0]

print("Original:", random_numbers)
print("%:", meth1)
print("np.floor:", meth2)
print("astype:", meth3)
print("np.trunc:", meth4)
````

#### 31. Ixtiyoriy sonlardan tashkil topgan massiv yarating, unga o'xshash qiymatlar kiriting. So'ng, o'xshash qiymatlarni topuvchi dastur tuzing.


`ðŸ’¡Ishora` np.bincount, argmax

````python
orig_array = np.array([5, 2, 8, 3, 5, 7, 8, 2, 3, 6, 7, 5, 8, 2])

# O'xshash qiymatlarni topish
counts = np.bincount(orig_array)
duplicates = np.where(counts > 1)[0] # nollarni olib tashlaymiz
duplicates
````
