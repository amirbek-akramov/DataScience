```python
    import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/housing_LR.csv')
df.head()
```

```python
df['district'].value_counts()
```

```python
housing = df[df.district=='Чиланзарский']
housing.head()
```

```python
X = housing['size']
X
```

```python
y = housing['price']
y
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=housing, x='size', y='price')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=housing, x='size', y='price', line_kws={"color": "red"})
plt.show()
```

```python
import numpy as np

Xmean = np.mean(X)
Ymean = np.mean(y)
```

```python
theta1 = sum((X-Xmean)*(y-Ymean)) / sum((X-Xmean)**2)
print(theta1)
```

```python
theta0 = Ymean - theta1*Xmean
print(theta0)
```

```python
x_test = housing.sample(10, random_state=42)['size'].to_numpy()
print(x_test)
y_test = housing.sample(10, random_state=42)['price'].to_numpy()
print(y_test)
```

```python
y_predict = theta0 + theta1*x_test
print(y_predict)
```

```python
MAE = np.mean(np.abs(y_predict-y_test))/len(y_test)
print(MAE)
```

```python
RMSE = np.sqrt(np.mean((y_predict-y_test)**2))
print(RMSE)
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(housing, test_size=0.10, random_state=42)
```

```python
from sklearn import linear_model
LR_model = linear_model.LinearRegression()

x_train = np.asanyarray(train_set[['size']])
y_train = np.asanyarray(train_set[['price']])

# Modelni tarbiyalaymiz (train)
LR_model.fit(x_train, y_train)

#theta1 va theta0 koeefisentlarni ajratib olamiz 
theta1 = LR_model.coef_[0][0]
theta0 = LR_model.intercept_[0]
print (f'{theta1=}')
print (f'{theta0=}')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=housing, x='size', y='price')
plt.plot(x_train, theta1*x_train + theta0, '-r')
plt.xlabel("Size")
plt.ylabel("Price")
```

```python
x_test = np.asanyarray(test_set[['size']])
y_test = np.asanyarray(test_set[['price']])
```

```python
y_predict = LR_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

MAE = mean_absolute_error(y_test, y_predict)
RMSE = np.sqrt(mean_squared_error(y_test, y_predict))
print(f"{MAE=}")
print(f"{RMSE=}")
```
