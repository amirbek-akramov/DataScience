```python
import numpy as np
```

```python
arr = np.arange(10)
arr
```

```python
#@title Universal functions - [sqrt] [square] [exp] [log] [modf] [sign] [isnan]
```

```python
#@title sqrt - roots of numbers
np.sqrt(arr)
```

```python
#@title square - square of numbers in array
np.square(arr)
```

```python
#@title exp - exponential
np.exp(arr)
```

```python
#@title log - logorithms
np.log(arr)
```

```python
#@title modf - modifier = 3.2=> 3 & 2
arr1 = np.random.rand(6)
arr1
```

```python
array, arra = np.modf(arr1)
print(array)
print(arra)
```

```python
#@title sign
arr2 = np.array([-1.35, 345, -634, 5, -7, 6, 6,-7])
print(arr2)
print(np.sign(arr2))
```

```python
#@title isnan - is there any nan values
np.isnan(arr)
```
